```python
import ffta
import pycroscopy as px
import pyUSID as usid
import numpy as np
from matplotlib import pyplot as plt
```

```python
# input file paths manually here
ibw_file = r'E:/Data/20190107 - BAPI trEFM/FF2_128_7V.ibw'
ff_folder = r'E:\Data\20190107 - BAPI trEFM\FF02_128x64_455nm_7V_400mA'
```

```python
h5_path, parm_dict, h5_avg = ffta.hdf_utils.load_hdf.loadHDF5_ibw(ibw_file_path=ibw_file, ff_file_path=ff_folder, 
                                                                  verbose=True, average=True)
```

```python
# SVD
h5_svd = ffta.analysis.svd.FF_SVD(h5_avg)
```

```python
# Plot abundance maps correctly
h5_U = h5_svd['U']
h5_V = h5_svd['V']
h5_S = h5_svd['S']

num_rows = parm_dict['num_rows']
num_cols = parm_dict['num_cols']

abun_maps = np.reshape(h5_U[:,:25], (num_rows, num_cols,-1))
eigen_vecs = h5_V[:16, :]

num_abun_rows = 4
num_abun_cols = 4
fig, a = plt.subplots(nrows=num_abun_rows, ncols=num_abun_cols, figsize=(10,10))

tt = np.transpose(abun_maps, (2, 0, 1))

# The pycroscopy plot wrapper doesn't work in Jupyter
for r in range(num_abun_rows):
    for c in range(num_abun_cols):
        
        a[r][c].imshow(tt[:][:][r*num_abun_rows+c], cmap='inferno')
```

```python
# Rebuild
clean_components = [0,1,2,3,4] # change based on SVD

h5_rb = ffta.analysis.svd.FF_SVD_filter(h5_avg, clean_components)
```

```python
# The actual FF-trEFM processing. 
# Note that plotting does not currently work in Jupyter...
tfp_rb, shift_rb, inst_freq_rb, h5_if = ffta.hdf_utils.analyze_h5.process(h5_svd.file, ref=h5_rb.name)
```

```python
# Save to file
ffta.hdf_utils.analyze_h5.save_CSV_from_file(h5_if.file, h5_if.parent.name)
```
